```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('clean_temp.csv', parse_dates=True, index_col=0)
df.head()
```

## Alternative: Automatic analysis / decomposition 

```python
sd = seasonal_decompose(df['TEMP'], model='additive', period=365)
```

```python
# plt.rcParams['figure.figsize'] = (100, 80)
# sd.plot()
```

```python
plt.rcParams['figure.figsize'] = (100, 16)
sd.trend.plot()
```

```python
sd.seasonal.plot()
```

```python
sd.resid.plot()
```
